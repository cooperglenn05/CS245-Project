```python
import sys
import re #USE REGULAR EXPRESSIONS TO AID IN FLEXIBLE EVALUATION OF FINAL VALUE ([a-zA-Z] will cover all alphabetic values) DO the same for my operators
print("Please enter an expression in the format:\n\n (a and b) -> c"
+ "\n\nThe and operator is represented as '^', the or as '/',\n the if-then, as '>', the if-and-only-if as '<>', and negation as '~'\n\n")
expression = input()

#regx = /\s/
#Messed with final truth value last time,
#In the future: Get the truth column program to function despite number of variables

numOfVariables = 0
var = ''
operator = ''
i = 0

if '^' in expression or 'and' in expression or '&&' in expression:
  operator += '^'
  expression = expression.replace('and', '^')
  expression = expression.replace('&&', '^')
if '//' in expression or 'or' in expression:
  operator += '/'
if '->' in expression:
  operator += '->'
if '<>' in expression:
  operator += '<>'
if '~' in expression:
  operator += '~'
for char in expression:
  i += 1
  if(char.isalpha()):
    numOfVariables += 1
    var += char
  elif(char == '(' or char ==')' or char == ' '):
    continue
  else:
    print("Invalid expression: Try again with a valid expression")
    sys.exit()

#Code finds numOfVariables/numOfRows
numOfRows = 2**int(numOfVariables)

# Initialize truthTable with header
truthTable = [[]]
for i in range(numOfVariables + 1):
  truthTable[0].append('')
for i in range(numOfVariables + 1):
  if(i == 0):
    truthTable[0][i] = var[i]
  elif(i < numOfVariables):
    if(var[i] == ''):
      break
    truthTable[0][i] = var[i]
  else:
    truthTable[0][i] = expression

#Fill out size of table
for _ in range(numOfRows):
  truthTable.append([''] * (numOfVariables + 1))


#The following code automatticall populates truth tables for 3 variables DO MORE RESEARCH ON PATTERNS FOR MORE VARIABLES
#THIS CODE IS NOW HIGHLY MODULAR AND NEW TRUTH TABLE STATEMENTS CAN BE ADDED HERE
currColumn = 0
j = 0
for i in range(numOfVariables):
  if(i == 0):
    for x in range(numOfRows):
      if(j < (numOfRows / 2)):
        j += 1
        truthTable[x+1][currColumn] = "T"
      else:
        j += 1
        truthTable[x+1][currColumn] = "F"
  elif(i == 1):
    currColumn = 1
    j = 0
    for x in range(numOfRows):
      if(x % 2 == 0):
        j += 1
        truthTable[x+1][currColumn] = "T"
      else:
        j += 1
        truthTable[x+1][currColumn] = "F"
  elif(i == 2):
    currColumn = 2
    j = 0
    if(numOfVariables == 3):
      for x in range(numOfRows):
        if(j < 2):
          j += 1
          truthTable[x+1][currColumn] = "T"
        elif(j < 4):
          j += 1
          truthTable[x+1][currColumn] = "F"
        elif(j < 6):
          j += 1
          truthTable[x+1][currColumn] = "T"
        else:
          j += 1
          truthTable[x+1][currColumn] = "F"


#Final truth table value (I.E. WHERE THE ACTUAL PROCESSING HAPPENS)


for i in range(1, numOfRows + 1): # Start from 1 to skip header row
  finalValue = ''
  for j in range(numOfVariables + 1):
    if (j < numOfVariables): # Process variable values
        finalValue += truthTable[i][j]
    else: # Assign the final value
      if(operator == '^'): #THESE WORK FOR FINAL VALUE OF ONLY SIMPLE ONE OPERATOR EXPRESSIONS (MAKE THEM WORK BASED ON THE TABLE, NOT THE FINALVALUE)
        for char in finalValue:
          if(char == 'F'):
            truthTable[i][j] = 'F'
            break
          else:
            truthTable[i][j] = 'T'
      elif(operator == '~'):
        for char in finalValue:
          if(char == 'T'):
            truthTable[i][j] = 'F'
            break
          else:
            truthTable[i][j] = 'T'
      elif(operator == '/'):
        for char in finalValue:
          if(char == 'T'):
            truthTable[i][j] = 'T'
            break
          else:
            truthTable[i][j] = 'F'
      elif(operator == '->'):
        if(finalValue[0] == 'T' and finalValue[1] == 'F'):
          truthTable[i][j] = 'F'
          break
        else:
            truthTable[i][j] = 'T'
      elif(operator == '<>'):
        if(finalValue[0] == finalValue[1]):
          truthTable[i][j] = 'T'
          break
        else:
            truthTable[i][j] = 'F'

#Displays the truth table
for row in truthTable:
    print(row)


print(operator)
```
